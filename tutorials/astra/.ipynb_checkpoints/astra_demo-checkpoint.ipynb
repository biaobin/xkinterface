# Define functions and variables

````python
afs_data = r'\\afs\ifh.de\group\pitz\data'

workdir = os.path.join(afs_data, 'lixiangk', 'sim', 'python', 'interface', 'tutorials', 'astra')
os.chdir(workdir)

I2B = lambda I: -(0.0000372+0.000588*I)
B2I = lambda B: (-B-0.0000372)/0.000588

field_maps = os.path.join('.', 'field-maps')

BSA = 1.0
sigma_x = sigma_y = BSA/4.

phi_gun, phi_booster = 0, 0

MaxE_gun = 60
MaxE_booster = 12.

Imain = 372; MaxB = I2B(Imain)
````

# Generate input files for `Generator` and `Astra`

`Generator` uses only one namelist called `Input`. Therefore, one just needs to give all the variables and their values when initializing it. The name of variables are case insensitive.

````python
# Define Input module
generator = Generator1(FNAME = 'beam.ini', IPart = 50000, Species = 'electrons', Q_total = -4.,\
                      Ref_Ekin = 0.0e-6, LE = 0.55e-3, dist_pz = 'i',
                      Dist_z = 'p', Lt = 21.5e-3, rt = 2e-3, Cathode = True,\
                      Dist_x = 'r', sig_x = sigma_x, Dist_px = 'g', Nemit_x = 0,\
                      Dist_y = 'r', sig_y = sigma_x, Dist_py = 'g', Nemit_y = 0)
````

After the initialization, one can update the values of variable by calling the `Module.set` method. If the specified variable doesn't exist, this method will add the new variable.

````python
generator.set(Q_total = -0.1, Ipart = 10000)
print(generator.output)
````

`Astra` has more than one namelists, therefore one needs to initialize them one by one. Similarly, one can update any variable with the `Namelist.set` method. 

Note that in some namelist, one variable can be assigned several values, for instance, `MaxE` in `Cavity`. In such a case, the variable must be assigned with a `list`, e.g., `MaxE = [60, 12]`.

````python
# Define namelists for Astra
newrun = Namelist('Newrun', Run = 1, Head = 'PITZ beam line simulation',
                Distribution = 'beam.ini', Auto_Phase = True, Track_All = True,
                check_ref_part = False, Lprompt = False)
newrun.set(Run = 1, Track_All = False)

charge = Namelist('Charge', LSPCH = True, Nrad = 50, Nlong_in = 50, N_min = 10,
                Max_scale = 0.05, Max_count = 20)

gun_profile = os.path.join(field_maps, 'gun42cavity.txt')
booster_profile = os.path.join(field_maps, 'CDS14_15mm.txt')
File_Efield = [gun_profile, booster_profile]

cavity = Namelist('Cavity', LEfield = True, File_Efield = File_Efield,
                MaxE = [MaxE_gun, MaxE_booster], C_pos = [0., 2.675],
                Nue = [1.3, 1.3], Phi = [phi_gun, phi_booster])

solenoid_profile = os.path.join(field_maps, 'gunsolenoidsPITZ.txt')
solenoid = Namelist('Solenoid', LBfield = True, File_Bfield = [solenoid_profile],
                  MaxB = MaxB, S_pos = [0.])

output = Namelist('Output', Zstart = 0, Zstop = 5.28, Zemit = 200, Zphase = 2, RefS = True,
                EmitS = True, PhaseS = True, TrackS = True, LandFS = True)

aperture_profile = os.path.join(field_maps, 'app1.txt')
aperture = Namelist('Aperture', LApert = True, File_Aperture = [aperture_profile])

quadrupole = Namelist('Quadrupole', Lquad = True)
````

Afterwards, one can create an `Astra` instance and add all the defined modules to it by a `list`. Finally, call the `write` method to output files for simulations.

````python
astra = Astra()
astra.add_modules([newrun, charge, cavity, solenoid, output, aperture])

# Write inputs file
generator.write('gen.in')
astra.write('ast.in')
````

## Parameter scan

There is an example python script for doing parameter scan in the folder `interface/tutorials/astra/ParaScan2024.py`. 

For this, one needs to define a function which wraps the above codes generating `Generator` and `Astra` input files. An input array is used to provide the functions the parameters one wants to vary/study.

````python
#%% Define goal function for Astra
def obj_ParaScan(x, *args):
    
    '''
    Created on Nov 17, 2019
    Simulation from photocathode via EMSY1 at 5.277 m to 20 m
    Goal function is a combination of average emittance and correlated energy spread exterpolated at undulator center, 29.25 m
    Laser spot fixed at 4 mm, gun phase at MMMG
    Variables to be optimized: gun and booster phases, solenoid current
    Parameters:
      x: an array or list of the variables to be optimized
    Returns:
      energy: the quantity that is envisoned as the "energy" of the sample
    '''
    
    Ipart = 250000
    Q_total = -x[0]/1e3 # to nC
    
    BSA = x[1]
    
    FWHM = 7e-3 # for Gaussian
    #FWHM = 20e-3 # for Flattop
    #FWHM = 250e-6 # for Pharos shortest pulse
    
    # uniform distribution
    sigma_x = sigma_y = BSA/4.
    C_sigma_x = C_sigma_y = 0
    
    sigma_x = sigma_y = 0.83 #0.96 # Gaussian truncated
    C_sigma_x = C_sigma_y = BSA/2./sigma_x
    
    phi_gun, phi_booster = x[3], x[5]
    MaxE_gun = get_MaxE_gun(phi_gun, 6.3)
    #MaxE_gun = x[2]
    
    MaxE_booster = get_MaxE_booster(MaxE_gun, phi_gun, phi_booster, 17)
    #MaxE_booster = x[4]
    
    Imain = x[6]
    MaxB = I2B(Imain)
    
    field_maps = os.path.join('.', 'field-maps')
    
    #Distribution = 'beam.ini'
    Distribution = f'BSA-{BSA:.1f}mm.ini'
    #Distribution = 'ast.0528.011'
    generator = Generator1(FNAME = Distribution, IPart = Ipart, Species = 'electrons', Q_total = Q_total,
                           Ref_Ekin = 0.0e-6, LE = 0.55e-3*1, dist_pz = 'i',
                           Dist_z = 'g', sig_clock = FWHM/2.355, Cathode = True,
                           Dist_x = '2', sig_x = sigma_x, Dist_px = 'g', Nemit_x = 0,
                           Dist_y = '2', sig_y = sigma_y, Dist_py = 'g', Nemit_y = 0,
                           C_sig_x = C_sigma_x, C_sig_y = C_sigma_y) # Gaussian by default
    
    #generator.set(Ref_Ekin = 1e-6, LE = 0, dist_pz = 'g', Nemit_x = 0.17, Nemit_y = 0.18)
    #generator.set(Dist_x = 'r', sig_x = BSA/4.0, Dist_y = 'r', sig_y = BSA/4.0) # Transverse uniform
    #generator.set(Dist_z = 'p', Lt = FWHM, Rt = 1e-3) # Temporal flattop
    
    Run = 1
    newrun = Module('Newrun', Run = Run, Head = 'PITZ beam line simulation', Distribution = Distribution, CathodeS = True,
                    Auto_Phase = True, Track_All = True, check_ref_part = False, Lprompt = False, Max_step=200000)
    #newrun.set(Xoff = 0.5, Yoff = 0.5)
    #newrun.set(Qbunch = Q_total)
    #newrun.set(Run = 1, XYrms = sigma_x)
    
    charge = Module('Charge', LSPCH = True, Lmirror = True, Nrad = 40, Nlong_in = 50, N_min = 50, Max_scale = 0.05, Max_count = 20)
    charge.set(N_min= 50)
    #charge.set(L2D_3D = True, Z_TRANS = 5, NXF = 32, NYF = 32, NZF = 32, min_grid_trans = 0.03e-3)
    #charge.set(LSPCH3D = True, NXF = 32, NYF = 32, NZF = 32)
    #charge.set(LSPCH = False)
    
    cavity = Module('Cavity', LEfield = True, File_Efield = 
                    [field_maps+os.sep+gun_profile, field_maps+os.sep+'CDS14_15mm.txt'],
                    MaxE = [MaxE_gun, MaxE_booster], C_pos = [0., 2.675], Nue = [1.3, 1.3], Phi = [phi_gun, phi_booster])
    
    soleno = Module('Solenoid', LBfield = True, File_Bfield = [field_maps+os.sep+'gunsolenoidsPITZ.txt'],
                    MaxB = [-MaxB], S_pos = [0.], S_xrot = [0*1e-3], S_yrot = [0])
    
    Zstart, Zstop = 0, 5.28
    Zemit = int((Zstop-Zstart)/0.01)
    output = Module('Output', Zstart = Zstart, Zstop = Zstop, Zemit = Zemit, Zphase = 1, RefS = True, EmitS = True,
                    PhaseS = True, TrackS = False, LandFS = True, C_EmitS = True, LPROJECT_EMIT = True,
                    LOCAL_EMIT = False, Screen = [5.28])
    #output.set(Zstop = 0.5, Zemit = 50) # emission
    
    apertu = Module('Aperture', LApert = True, File_Aperture = [field_maps+os.sep+'app.txt'])
    
    astra = Astra()
    astra.add_modules([newrun, charge, cavity, soleno, output])
    
    direc = str.format('Q-%.2fpC-D-%.2fmm-E1-%.2fMV_m-phi1-%.2fdeg-E2-%.2fMV_m-phi2-%.2fdeg-I-%.2fA' %\
                       (np.abs(x[0]), BSA, MaxE_gun, phi_gun, MaxE_booster, phi_booster, Imain))
    
    ### Generating input files for Farm
    try:
        os.mkdir(direc)
    except Exception as err:
        print(err)
        
    job_name = direc+'-%03d' % Run
    #job_name = 'myjob.'+('BSA-%.2fmm-I-%.2fA' % (x[0], x[3]))
    gen_name = 'gen' #+`Run`
    ast_name = 'ast' #+`Run`
    
    generator.write(direc+os.sep+gen_name+'.in')
    astra.write(direc+os.sep+ast_name+'.in')
    #astra.qsub(job_name, ast_name, gen_name, direc)
    #os.system('qsub '+job_name+'.sh')
    
    astra.submit(job_name, ast_name, gen_name, direc)
    
    return
    ### End of generating input files for Farm

    ### Or if one wants to run from within python, e.g., for multiobjective optimization
    os.system('mkdir -p '+direc)
    os.chdir(direc)
    
    generator.write('gen.in')
    astra.write('ast.in')
    
    os.system('generator gen.in 2>&1 | tee gen.log')
    os.system('astra ast.in 2>&1 | tee ast.log')

    try: 
        os.chdir(direc)
        try:
            fname = 'ast.0528.001'
            diag = BeamDiagnostics(fname = fname, energy = False)
        except:
            fname = 'ast.0527.001'
            diag = BeamDiagnostics(fname = fname, energy = False)
            
        res = list(x) + list(diag.x)
        os.chdir('..')

        with open('ParaScan@5.28m.dat','a') as f_handle:
            np.savetxt(f_handle,np.atleast_2d(res),fmt='%14.6E')

        obj = [diag.nemit_x, diag.std_z]
    except:
        # if no output beam file, then set the objective to 10000
        os.chdir('..')
        print('Error: not accelerated well!')
        obj = [999, 999]
    ### End of function evaluation for multiobjective optimization
    
    return obj
````

````python
#%% Parameter scan using Astra
# Charge, pC
var0 = [1000]

# BSA, mm
var1 = [3.5]

# Gun phase, degree
var2 = np.linspace(-9, 0, 4)
var2 = [0]

# Booster phase, degree
var3 = np.linspace(-36, -16, 6)
#var3 = [-28, -24, -20]

# Solenoid, A
var4 = np.linspace(360, 374, 8)

combi = np.array([[v0, v1, 57.55, v2, 14.5, v3, v4] for v0 in var0 for v1 in var1 for v2 in var2 for v3 in var3 for v4 in var4])

for x in combi[:1]:
    print(x)
    obj_ParaScan(x)
    #post_ParaScan(x)
    pass
````
